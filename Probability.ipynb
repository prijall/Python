```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import scipy.special as spsp
```

```python
n=30
p=0.03
q=1-p

probBinomial=[]
for k in range(n+1):
    prob1=spsp.comb(n, k) * p**k * q**(n-k)
    probBinomial.append(prob1)

mu=n*p
sigma=np.sqrt(n*p*q)
 
X=np.arange(0, n, 0.01)
prob2=(1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-((X-mu)**2)/(2*sigma**2))


plt.plot(probBinomial, label="Binomial")
plt.plot(X, prob2, label='Normal')
plt.legend()
plt.xlim(-5, 5)
```
