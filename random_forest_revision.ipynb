```python
import opendatasets as od
import pandas

od.download(
    "https://www.kaggle.com/datasets/dhgupta/heartdiseasedata")
```

```python
import pandas as pd

file=('heartdiseasedata/heart_v2.csv')
df=pd.read_csv(file)
```

```python
df.head()
```

```python
#@ Importing necessaries libraries and dependencies:
import numpy as np
import matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
```

```python
X=df.drop('heart disease', axis=1)
y=df['heart disease']
```

```python
y
```

```python
#@ train-test split:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X, y, train_size=0.8, random_state=22)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(n_estimators=100, max_depth=5, n_jobs=-1, oob_score=True, random_state=22)


rf.fit(X_train , y_train)
```

```python
rf.oob_score_
```
